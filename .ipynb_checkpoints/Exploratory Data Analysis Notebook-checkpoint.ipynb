```python
def query_dataset(dataframe,column,expression):
    """
    args:
    dataframe is the dataframe that the column will be extracted from to do the boolean expression 
    (string)column : is the column name of the dataset
    expression: is the boolean expression 
    
    return:
       a dataframe/dataset that satisfies the boolean expression
    
    """
    dataset=dataframe[dataframe[column] ==expression]
    return dataset
```

```python
def plot_dist(row,col, data):
    fig1, ax1 = plt.subplots(row, col, figsize=(40, 30))
    for i, feature in enumerate(data.columns):
        row = i // num_cols
        col = i % num_cols
        sns.histplot(df[feature], kde=True, ax=ax1[row, col])
        
        ax1[row, col].set_title(f"Distribution of {feature}", fontsize="25")

    fig1.tight_layout()
    
```

## Plotting Distribution plots before dropping outliers

```python
plot_dist(5,4,df)
```

```python
def remove_outliers_by_mad(data, columns, confidence_level=3.5):


    data_filtered = data.copy()
    
    for column in columns:

        median = data_filtered[column].median()
        mad = np.median(np.abs(data_filtered[column] - median))
        
        # Calculate the modified z-score
        modified_z_scores = 0.6745 * (data_filtered[column] - median) / mad
        
        # Remove the outliers based on the threshold
        data_filtered = data_filtered[abs(modified_z_scores) <= confidence_level]
    
    return data_filtered

```

```python
    df = remove_outliers_by_mad(df, numerical_cols,confidence_level=3.5 )
```
